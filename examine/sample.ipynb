```python
# 生成AI系のインポート
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()

llm=ChatOpenAI(model="gpt-4.1-nano")
prompt=ChatPromptTemplate.from_messages([
    ("system", "あなたは優秀なデータ分析者です。"),
    ("human", "以下のデータを分析してください。\n"
             "データの内容は{data}です。\n"
             "この表には複数の表構造が含まれています。\n"
            #  "【重要指数】、【PL】、【指標】の各表の行数と列数を教えてほしい。\n"
            #  "回答形式は「重要指数：〇×〇, PL：〇×〇、指標：〇×〇」のようにしてください。\n"
            #  "回答形式を必ず守ってください。\n"
             "重要指数の表に含まれる行タイトルをすべて教えてください。\n"
             "大分類・中分類を含むような場合は、最も小さい分類まで表示してください。\n"
             "小分類とは、◆重要指数の表構造でいうと、「レギュラー」「アドバンス」のような分類のことです。\n"
             "回答のみを表示してください。\n")
])

chain=prompt | llm

result=chain.invoke({"data":df.to_dict()})
print(result.content)
```
